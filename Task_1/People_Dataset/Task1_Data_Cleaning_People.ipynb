```python
df = df.drop(columns=["Index"], errors="ignore")

df = df.rename(columns={
    "User Id": "user_id",
    "First Name": "first_name",
    "Last Name": "last_name",
    "Sex": "gender",
    "Email": "email",
    "Phone": "phone",
    "Date of birth": "dob",
    "Job Title": "job_title"
})

df.head()
```

```python
df.isna().sum()
```

```python
(df == "").sum()
```

```python
df.apply(lambda col: col.str.isspace().sum() if col.dtype=="object" else 0)
```

```python
df["dob"] = pd.to_datetime(df["dob"], errors="coerce")

df[["dob"]].head()
```

```python
df["dob"].isna().sum()
```

```python
# Convert to string, remove all non-digit characters
df["phone"] = df["phone"].astype(str).str.replace(r"\D", "", regex=True)

df["phone"].head()
```

```python
(df["phone"] == "").sum()
```

```python
# Standardize name + job title formatting
text_cols = ["first_name", "last_name", "job_title"]

for col in text_cols:
    df[col] = df[col].str.strip().str.title()

# Standardize gender text
df["gender"] = df["gender"].str.strip().str.lower().replace({
    "m": "male",
    "f": "female"
})

df[["first_name", "last_name", "gender", "job_title"]].head()
```

```python
df["user_id"].duplicated().sum()
```

```python
df["user_id"].isna().sum()
```

```python
df.info()
```

```python
df.to_csv("cleaned_people_data.csv", index=False)

print("Dataset cleaned and saved successfully!")
```
