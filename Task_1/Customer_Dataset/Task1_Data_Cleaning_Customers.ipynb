```python
# Drop the Index column
df_cust = df_cust.drop(columns=["Index"], errors="ignore")

# Rename columns for consistency
df_cust = df_cust.rename(columns={
    "Customer Id": "customer_id",
    "First Name": "first_name",
    "Last Name": "last_name",
    "Company": "company",
    "City": "city",
    "Country": "country",
    "Phone 1": "phone_primary",
    "Phone 2": "phone_secondary",
    "Email": "email",
    "Subscription Date": "subscription_date",
    "Website": "website"
})

df_cust.head()
```

```python
df_cust.apply(lambda col: col.str.isspace().sum() if col.dtype=="object" else 0)
```

```python
df_cust["subscription_date"] = pd.to_datetime(
    df_cust["subscription_date"],
    errors="coerce"
)

df_cust[["subscription_date"]].head()
```

```python
df_cust["subscription_date"].isna().sum()
```

```python
# Clean Primary Phone
df_cust["phone_primary"] = (
    df_cust["phone_primary"]
    .astype(str)
    .str.replace(r"\D", "", regex=True)
)

# Clean Secondary Phone
df_cust["phone_secondary"] = (
    df_cust["phone_secondary"]
    .astype(str)
    .str.replace(r"\D", "", regex=True)
)

df_cust[["phone_primary", "phone_secondary"]].head()
```

```python
(df_cust["phone_primary"] == "").sum(), (df_cust["phone_secondary"] == "").sum()
```

```python
# Title-case selected text columns
cols_title = ["first_name", "last_name", "company", "city", "country"]

for col in cols_title:
    df_cust[col] = df_cust[col].str.strip().str.title()

# Lowercase emails (best practice)
df_cust["email"] = df_cust["email"].str.strip().str.lower()

# Standardize Customer ID to uppercase
df_cust["customer_id"] = df_cust["customer_id"].str.strip().str.upper()

df_cust[["customer_id","first_name","last_name","company","city","country","email"]].head()
```
