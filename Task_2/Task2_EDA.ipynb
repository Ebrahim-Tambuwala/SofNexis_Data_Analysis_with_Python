```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
# Visualizing the age distribution of passengers to understand the overall age spread and identify common age groups
sns.histplot(df["Age"], kde=True, bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Plotting the fare distribution to observe price variation and detect any high-fare outliers among passengers
sns.histplot(df["Fare"], kde=True, bins=30)
plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,5))
# Counting the number of male and female passengers to understand the gender distribution on the ship
sns.countplot(x="Sex", data=df)
plt.title("Passenger Count by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,5))
# Comparing survival counts between males and females to see which gender had higher chances of survival
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival Count by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,5))
# Analyzing how survival varies across passenger classes (1st, 2nd, 3rd) to identify which class had better survival chances
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Comparing the age distribution of survivors vs non-survivors to see which age groups had higher chances of survival
sns.kdeplot(df[df["Survived"]==1]["Age"], label="Survived", fill=True)
sns.kdeplot(df[df["Survived"]==0]["Age"], label="Did Not Survive", fill=True)
plt.title("Age Distribution by Survival")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Comparing fare ranges of survivors and non-survivors to understand whether passengers who paid higher fares had better survival chances
sns.kdeplot(df[df["Survived"]==1]["Fare"], label="Survived", fill=True)
sns.kdeplot(df[df["Survived"]==0]["Fare"], label="Did Not Survive", fill=True)
plt.title("Fare Distribution by Survival")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Using a boxplot to detect extreme fare values and identify high-paying passengers as potential outliers
sns.boxplot(x=df["Fare"])
plt.title("Fare Outliers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Visualizing age outliers to check for unusually young or elderly passengers compared to the general age distribution
sns.boxplot(x=df["Age"])
plt.title("Age Outliers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Examining how males and females are distributed across the three passenger classes to understand class–gender patterns on the ship
sns.countplot(x="Pclass", hue="Sex", data=df)
plt.title("Gender Distribution Across Passenger Classes")
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Analyzing survival rates by combining passenger class and gender to see which groups had the highest or lowest chances of survival
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df)
plt.title("Survival Rate by Class & Gender")
plt.show()
```

```python
# Creating a new feature 'FamilySize' to study whether travelling alone, in small families, or in large groups affected survival chances
df["FamilySize"] = df["Siblings/Spouses Aboard"] + df["Parents/Children Aboard"] + 1

plt.figure(figsize=(8,5))
# Visualizing how survival rate changes with different family sizes
sns.barplot(x="FamilySize", y="Survived", data=df)
plt.title("Survival Rate by Family Size")
plt.show()
```

## 📄 EDA Report

### 1. Dataset Overview: Size, features, missingness

* **Rows:** 887
* **Columns:** 8
* **Features include:** Survival status, passenger class, gender, age, fare, and family members onboard
* **Data Types:** Mix of numerical and categorical
* **Missing Values:** **Detected in Age and Embarked; handled via median/mode imputation** 

### 2. Key Statistics: Central tendencies, spreads

* **Average Age:** ~29.5 years
* **Median Age:** 28 years
* **Average Fare:** ~£32
* **Median Fare:** ~£14.45
* **Age Range:** 0.42 to 80 years
* **Fare Range:** 0 to £512
* **Key distributions:**
* Majority of passengers were **20–40 years old**
* Most travellers were in **3rd class**
* More **men (573)** than **women (314)** were onboard

### 3. Critical Visuals: 4–6 annotated plots

The following plots were analyzed:

1. **Age Distribution Histogram + KDE:** Shows most passengers were young adults; small number of elderly.
2. **Fare Distribution Histogram:** Right-skewed distribution; a few very high-fare passengers.
3. **Survival by Gender (Countplot):** Significantly more women survived than men.
4. **Survival by Class (Countplot):** 1st-class passengers survived most; 3rd-class the least.
5. **Survival Rate by Class × Gender (Barplot):** Women in 1st & 2nd class had the highest survival; men in 3rd class had the lowest.
6. **Family Size vs Survival (Barplot):** Small families (2–4 people) survived the most; solo & large-family travellers survived less.

### 4. Anomaly Handling (Outliers)

**Observation:** Identified extreme fares (> £500) via Z-score analysis. 

**Decision:** **Kept for analysis.** These represent legitimate 1st-class suites and are valuable for understanding wealthy passenger survival. 

### 5. Hypotheses Tested

**○ H1: Higher-class passengers had better survival**

**Result: ✅ Confirmed** 

* 1st-class survival rate was the highest
* 3rd-class survival rate was the lowest

**○ H2: Age impacted survival non-linearly**


**Result: ✅ Confirmed** 

* Children had higher survival probability
* Middle-aged and elderly survival rates were lower
* Indicates **non-linear age effect**
